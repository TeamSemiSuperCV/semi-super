# Stage 2 fine-tuning is run for five times with the same settings (below ) and the average test accuracy is used for reporting.

```python
# Run 1
model_folder = 'reborn_again_1pc_1'

ckpt_dir = 'gs://simclrv2-sb/reborn_again_500_1e4/saved_model/18500'
model_dir = 'gs://simclrv2-sb/'

class simclrCommand():
  def __init__(self, model_dir, model_folder, params={}):
    #default params
    self.params = {'train_mode': "finetune", 
          "mode": "train_then_eval",    
          "resnet_depth": 50,
          "width_multiplier": 2,
          "sk_ratio":0.0625,  
          "color_jitter_strength": 0.5,
          'model_dir': model_dir+model_folder,
          "data_dir": model_dir,
          "use_tpu": True}
    # add in or change params
    if len(params) > 0:
      for k,v in params.items():
        self.params[k]=v
  def compile_command(self):
    simclr_command = ['python run.py']
    for k,v in self.params.items():
      simclr_command.append(f'--{k}={v}')
    return (" ").join(simclr_command)
  def run_command(self):
    !{self.compile_command()}

params = {"train_epochs": 60,
          'warmup_epochs': 0,
          'checkpoint_epochs': 10,
          "train_batch_size": 32,
          'train_split': 'train_1pc', 
          "image_size": 224,   
          'dataset': "xray_reborn",
          "learning_rate": 0.001, 
          "learning_rate_scaling": 'sqrt',
          'weight_decay': 0, 
          "temperature": 0.1,
          "use_blur": False, 
          "area_range_min": 1,
          "eval_split": "validation",    
          "eval_per_loop": False,
          'fine_tune_after_block': -1,
          'zero_init_logits_layer': True,
          "checkpoint": ckpt_dir} 
slimsk2 = simclrCommand(model_dir, model_folder, params)
slimsk2.run_command()

slimsk2.params['mode'] = 'eval'
slimsk2.params['eval_split'] = 'test'

slimsk2.run_command()
```
